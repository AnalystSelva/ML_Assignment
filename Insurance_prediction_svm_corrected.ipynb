```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset.shape[0]
```

```python
dataset.shape[1]
```

```python
one_hot_encoding = pd.get_dummies(dataset,drop_first=True)
one_hot_encoding = one_hot_encoding.astype(int)
one_hot_encoding.head()
```

```python
one_hot_encoding.columns
```

```python
independent = one_hot_encoding[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent = one_hot_encoding[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
x_train , y_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVR
```

```python
regression = SVR(kernel="rbf", C=5000)
regression
```

```python
regression.fit(x_train,y_train)
```

```python
# it support only linear
#weight = regression.coef_
#weight
regression.n_support_
regression.support_
```

```python
bias = regression.intercept_
bias
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = regression.predict(x_test)
y_pred
```

```python
r2_score = r2_score(y_test, y_pred)
r2_score
```
